```python
# difining price function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#appliying function to dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

## Task

##### change the busiest day to busiest days

```python
df_ord_mer["orders_days_of_the_week"].value_counts(dropna=False)
```

```python
# 0 and 1 stand for saturday and sunday being the busiest days. 3 and 4 stand for wednesday and tuesday being the slowest days
days=[]

for value in df_ord_mer["orders_days_of_the_week"]:
  if value == 0 or value==1:
    days.append("Busiest day")
  elif value == 4 or value==3:
    days.append("Least busy")
  else:
    days.append("Regularly busy")
```

```python
#create the column
df_ord_mer["busiest_days"]=days
```

```python
#count the values of days
df_ord_mer["busiest_days"].value_counts(dropna=False)
```

```python
# checking busiest hour
df_ord_mer["time_of_day"].value_counts(dropna=False)
```

```python
### i will divide the result in 3 groups 

hours = []

for hour in df_ord_mer ["time_of_day"]:
  if hour in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append('Most orders')
  elif hour in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
# creating new column
df_ord_mer["busiest_period_of_day"]=hours
```

```python
#count of busiest hours of the day
df_ord_mer["busiest_period_of_day"].value_counts(dropna=False)
```
